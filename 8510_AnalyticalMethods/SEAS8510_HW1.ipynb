# 1 Plot Addtion and Subtraction of Vectors

```python
def plot_vector(start, end, color, vector):
    x0, y0 = start
    x1, y1 = end
    # Calculate the differences for vector
    dx = x1 - x0
    dy = y1 - y0
    plt.arrow(x0, y0, dx, dy, head_width=0.2, head_length=0.3, fc=color, ec=color, label=vector)

plt.figure(figsize=(5,4))
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(True)

plot_vector((0, 0), (1, 2), "blue", "V")
plot_vector((1, 2), (5, -4), "green", "W")
plot_vector((0, 0), (5, -4), "red", "V+W")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(True)

plot_vector((0, 0), (1, 2), "blue", "V")
plot_vector((0, 0), (4, -6), "green", "W")
plot_vector((4, -6), (1, 2), "red", "V-W")

plt.legend()

plt.show()
```

# 2 Calcuate the L2 Norm of a Vector

```python
# Let's define two arrays (one small 2 dimensions and larger 10 dimensions)

A = np.array([1,2])
B = np.array([4,-2, 0, 1, 3, 4, -6, -7, 0, 4])
print(A)
print(B)
```

```python
# To calculate norm we can make use the dot product of the vector with itself
def L2_norm(vector):
    return (np.sqrt(vector.dot(vector)))

print(L2_norm(A))
print(L2_norm(B))
```

```python
# Let's check our results vs np.linalg.norm()
print(np.linalg.norm(A))
print(np.linalg.norm(B))
```

# 3 Convert to a Unit vector any vector

```python
def norm_vector(vector):
    return vector / L2_norm(vector)

print("Normalized vector A", norm_vector(A))
print("Let's check the Norm of the normalized vector A: ", L2_norm(norm_vector(A)))
print()
print("Normalized vector B", norm_vector(B))
print("Let's check the Norm of the normalized vector B: ", L2_norm(norm_vector(B)))
```

```python
zero_vector = np.array([0,0,0])
print(norm_vector(zero_vector))
```

# 4 Create function to normalize to any magnitud


```python
def norm_m_vector(vector, m):
    return m * vector / L2_norm(vector)

#unit of measure to normalize to:
m = 5

print("Normalized vector A", norm_m_vector(A,m))
print("Let's check the Norm of the normalized vector A: ", L2_norm(norm_m_vector(A,m)))
print()
print("Normalized vector B", norm_m_vector(B,m))
print("Let's check the Norm of the normalized vector B: ", L2_norm(norm_m_vector(B,m)))
```

# 5 Transpose with a loop

```python
import numpy as np

def transpose_with_loops(matrix):
    rows, cols = matrix.shape
    
    # Initialize a new matrix with swapped dimensions
    transposed_matrix = np.zeros((cols, rows))
    
    # Loop over rows and columns to transpose the matrix
    for i in range(rows):
        for j in range(cols):
            transposed_matrix[j, i] = matrix[i, j]
    
    return transposed_matrix

# Example usage
matrix = np.array([[1, 2, 3], [4, 5, 6]])
transposed = transpose_with_loops(matrix)
print(matrix)
print()
print(transposed)
```
