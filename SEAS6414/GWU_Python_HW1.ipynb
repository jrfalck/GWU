```python
def calculate_npv(cash_flows, daily_discount_rate):
    npv = 0
    for day, cash_flow in enumerate(cash_flows):
        discounted_cash_flow = cash_flow / ((1 + daily_discount_rate) ** day)
        npv += discounted_cash_flow
    return npv

cashflowbuy = 450 * 550 * -1  #Is negative since we purchased the stock
cashflowsell = 450 * 950 #Is positive since we sold the stock



cash_flows =[0] * 651  # one more day to account for day 0


daily_discount_rate = 0.00015/100  # example daily discount rate
cash_flows[0] = cashflowbuy
cash_flows[650] = cashflowsell


npv = calculate_npv(cash_flows, daily_discount_rate)
print(f"The Net Present Value (NPV) is: {{:,.0f}}".format(npv))
```

```python
def evenodd_squareandroot(number):
    from math import sqrt
    
    if number % 2 == 0:
        evenvalue = True
        calcvalue = number ** 2
    else:
        evenvalue = False
        calcvalue = sqrt(number)
    
    return evenvalue,calcvalue

    
# Example
num = 10
even_check, calculated_value = evenodd_squareandroot(num)
print(f"The number {num} is even? {even_check}.")
print(f"The resulting calculation of number {num} is {calculated_value}.")
```

```python
def create_consecutive_ints(s1, s2, s3):
    
    combined_string = s1 + s2 + s3

    # Remove duplicates and sort
    unique_sorted_digits = ''.join(sorted(set(combined_string)))

    
    if unique_sorted_digits == '123456789':
        return unique_sorted_digits
    else:
        return "The strings do not form a sequence of consecutive integers from 1 to 9."

# Example usage
s1 = "12345"
s2 = "34567"
s3 = "789"
consec_ints = create_consecutive_ints(s1, s2, s3)
print(consec_ints)
```

```python
def handle_list(list):
    first = list[0]
    last = list[-1]
    sumvalues = first + last
    return sumvalues

M = [3,12,25,8,16,42,9,32,11]
sum_first_last = handle_list(M)
print(f"The sum of the first and last values of the list is: {sum_first_last}.")
```

```python
def get_middle_element(my_list):
    length = len(my_list)
    middle_index = length // 2
    if length % 2 == 0: 
        mid_element = "List has even number of elements"
    else: 
        mid_element = my_list[middle_index]
    
    new_list = my_list[:middle_index]
    
    return mid_element, new_list

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [1, 2, 3, 4]

mid_element, sub_list = get_middle_element(list1)
print(mid_element)
print(sub_list)

mid_element, sub_list = get_middle_element(list2)
print(mid_element)
print(sub_list)


```
