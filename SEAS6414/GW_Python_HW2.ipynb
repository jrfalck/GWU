```python
def create_dictionary(sentence):
 
    words = sentence.split()

    word_lengths = {word: len(word) for word in words}

    return word_lengths


sentence = "I love Machine Learning"
result = create_dictionary(sentence)
print(result)
```

```python
def count_substring(str1, str2):
    count = 0
    start = 0

    while start < len(str1):

        pos = str1.find(str2, start)
        
        if pos == -1:
            break

        count += 1
        start = pos + 1  

    return count

# Example usage
str1 = "coding is cool"
str2 = "co"
result = count_substring(str1, str2)
print(result)
```

```python
def remove_char(string):
    return [string[:i] + string[i+1:] for i in range(len(string))]

# Example usage
input_string = "Wale"
result = remove_char(input_string)
print(result)
```

```python
D = {"Jake": [99, 87, 91, 77], "Charlie": [100, 100, 99], "Ellen": [95, 70, 85, 100, 100]}

def calculates(scores_dict):
    average_scores_dict = {}

    for student, scores in scores_dict.items():
        average_score = sum(scores) / len(scores)
        average_scores_dict[student] = average_score

    return average_scores_dict

average_scores = calculates(D)
print(average_scores)
```

```python
import string

sentence = "Tom enjoyed reading books on philosophy. He often pondered the deeper meanings of life, especially during quiet nights."

def filter_words(sentence):
  
    words = sentence.split()

    long_words = []

    for word in words:
        cleaned_word = word.strip(string.punctuation)
        if len(cleaned_word) > 5:
            long_words.append(cleaned_word)

    return long_words


long_words_list = filter_words(sentence)
print(long_words_list)
```

```python
import string

def mycounter(sentence):
    sentence = sentence.lower()

    for char in string.punctuation:
        sentence = sentence.replace(char, '')

    words = sentence.split()
    vowels = 'aeiou'
    vowel_count = sum(1 for word in words if word[0] in vowels)

    return vowel_count

sentence = "During her summer vacation, Alice explored various historical sites and enjoyed local cuisines."

count = mycounter(sentence)
print("Number of words starting with a vowel:", count)
```

```python
def parse_string(term):
    result = []
    count = 1

    for i in range(1, len(term)):
        if term[i] == term[i-1]:
            count += 1
        else:
            result.append([int(term[i-1]), count])
            count = 1

    result.append([int(term[-1]), count])
    
    return result

def next_term(term):
    parsed = parse_string(term)
    return ''.join(str(count) + str(digit) for digit, count in parsed)

def read_it(init, n):
    term = init
    for _ in range(n):
        print(term)
        term = next_term(term)


init = "1"
n = 11
read_it(init, n)
```
